# Matplotlib : is a python libraryn that used to undersatnd the data by the graphical representation of the data

There are three types of analysis for matplotlib 
1.univariate analysis: to show the relation of a one particular columns
2.bivariate analysis: to show the relationship b/w two columns

There are different types of plots:
    1.scatter plot
    2.plot
    3.histogram
    4.bar plot
    5.count plot
    6.sub plot
    7.pie plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. scatter
x=[1,2,3,4]
y=[5,6,7,8]
plt.figure(figsize=(3,3))
plt.scatter(x,y,color='r',label='hi')
plt.legend(loc=0)
plt.show()
```

```python
##2.plot
x=range(1,50)
y=[value*3 for value in x]
plt.figure(figsize=(5,5))
plt.plot(x,y,color='b',label='hi')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(1,50)
plt.ylim(1,160)
plt.title('drawing the line')
plt.legend(loc=0)
plt.show()
```

```python
#3.hist
x=[1,2,3,4,5,66,67,88,90]
plt.hist(x,bins=5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('histogram for univariate analysis')
plt.show()
plt.show()
```

```python
#3.hist for bivariate analysis
x=[1,2,3,4,5,66,67,88,90]
y=[1,4,5,6,7,8,8,99,9]
plt.hist([x,y],bins=5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('histogram')
plt.show()
plt.show()
```

```python
##4.Bar plot:is used to show the label values in a column
a=['Java','Python','PHP','JavaScript','C#','C++']
x=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(a,x,color=(0.4, 0.6, 0.8, 1.0))
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
##5.pie chart : is used to display the proportions of the labels in the column
x=['Java','Python','PHP','JavaScript','C#','C++']
y=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.pie(y,labels=x,autopct='%1.1f%%',explode=(0,0.5,0,0,0,0))
plt.show()
```

```python
#6.subplot: is used to show the multiple plots in rows and columns data
x=[1,2,3]
y=[2,4,1]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('drawing the line')
plt.show()
x=[1,2,3]
y=[2,4,1]
a=[1,2,3]
b=[4,1,3]
plt.subplot(1,2,2)
plt.plot(x,y)
plt.plot(a,b,color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('drawing the line')
plt.show()
```

# Different types of plots for analysis of the matplotlib

```python
x=[1,2,3]
y=[2,4,1]
a=[1,2,3]
b=[4,1,3]
plt.plot(x,y)
plt.plot(a,b,color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('drawing the line')
plt.show()
```

```python
x=[1,2,3]
y=[2,4,1]
a=[1,2,3]
b=[4,1,3]
plt.plot(x,y)
plt.plot(a,b, color='red', linewidth = 5,  label = 'line2-dashed', linestyle='dashed')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('drawing the line')
plt.show()
```

```python

import matplotlib.pyplot as plt
# x axis values
x = [1,4,5,6,7]
# y axis values
y = [2,6,3,6,3]

# plotting the points 
plt.plot(x, y, color='red', linestyle='dashdot', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)
#Set the y-limits of the current axes.
plt.ylim(1,8)
#Set the x-limits of the current axes.
plt.xlim(1,8)

# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')

# giving a title to my graph
plt.title('Display marker')
# function to show the plot
plt.show()
```

```python
x=[1,4,5,6,7]
y=[2,6,3,6,3]
plt.plot(x,y,'r*',linestyle='dashdot',linewidth=3,markerfacecolor='blue',markersize=14)
plt.xlim(1,8)
plt.ylim(1,8)

plt.show()
```
